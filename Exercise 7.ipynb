## **Arithmetic Expressions**

Arithmetic expressions are basic mathematical operations that can be performed in Python.  
Here are some simple examples:

- **Addition:** `5 + 3`
- **Subtraction:** `10 - 4`
- **Multiplication:** `6 * 7`
- **Division:** `20 / 5`
- **Exponentiation:** `2 ** 3`
- **Modulus (remainder):** `10 % 3`

These expressions help us perform basic calculations in programming and mathematics.
