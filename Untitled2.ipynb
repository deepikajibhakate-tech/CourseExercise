## **Data Science Languages**

Here are some commonly used programming languages in Data Science:

- **Python** – Most popular for data analysis, ML, and visualization  
- **R** – Excellent for statistical computing  
- **SQL** – Used for querying and managing databases  
- **Julia** – High-performance computing and numerical analysis  
- **Scala** – Used with Apache Spark for big data processing  
- **MATLAB** – Used for numerical and scientific computing
